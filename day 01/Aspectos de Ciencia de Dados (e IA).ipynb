## Cronologia ⏱️

<ul>
    <li>1962: John Tukey descreveu um campo que chamou de "análise de dados" -- que se assemelha à atual ciência de dados;</li>
    <li>1974: Peter Naur propôs "ciência de dados" como um nome alternativo para "ciência da computação";</li>
    <li>1985: C. F. Jeff Wu usou o termo "ciência de dados" pela primeira vez como um nome alternativo para "estatística";</li>
    <li>1989: Gregory Piatetsky-Shapiro organiza e preside o primeiro workshop Knowledge Discovery in Databases (KDD) <br />-- renomeado em 1995 para Knowledge Discovery and Data Mining;</li>
    <li>2001: William S. Cleveland defendeu uma expansão da estatística além da teoria para áreas técnicas <br />-- porque isso mudaria significativamente a área, justificaria um novo nome: ciência de dados;</li>
    <li>2012: Thomas H. Davenport e D. J. Patil declararam "Cientista de Dados: O Trabalho Mais Sexy do Século XXI";</li>
    <li>2022: <a href='https://hbr.org/2022/07/is-data-scientist-still-the-sexiest-job-of-the-21st-century'>Is Data Scientist Still the Sexiest Job of the 21st Century?</a>
</ul>

### Atitude

<pre>Growth mindset</pre>

<img src='https://bettereveryday840.files.wordpress.com/2018/06/picture1.png' style='float:right;width:200px;'>

<ol>
    <li>A ‘zona de conforto’ é onde a maioria das pessoas passa a maior parte do tempo no trabalho. <br />O trabalho pode ser complicado ou complexo, mas não é difícil, desafiador ou novo;</li>
    <li>A 'zona de crescimento' é onde uma pessoa precisa se adaptar e aprender novas habilidades e, <br />embora seja desafiadora e desconfortável, ela pode progredir;</li>
    <li>A 'zona de pânico' é onde a situação é tão diferente ou estressante que <br />está além da capacidade de uma pessoa lidar.</li>
</ol>

<img src='https://bettereveryday840.files.wordpress.com/2018/06/picture11.png' style='float:left;width:350px;'>

quando você se expõe repetidamente, em pequenas doses, você pode aumentar sua tolerância

🚨 permite que pessoas lidem com maior estresse -- 

aumentando as zonas de conforto e crescimento

<img src='https://bettereveryday840.files.wordpress.com/2018/06/picture22.png' style='float:left;width:350px;'>

quando você não se expõe, você pode diminuir sua tolerância

🚨 torna pessoas menos capazes de lidar com choques repentinos -- 

reduzindo capacidade de desenvolver novas habilidades ou assumir novas responsabilidades

<img src='https://bettereveryday840.files.wordpress.com/2018/06/picture3.png' style='float:left;width:350px;'>

quando você se expõe demais, pode ser levado à zona de pânico

🚨 o estresse pode se manifestar em sintomas físicos ou mentais -- 

reduzindo capacidade de tomar decisão e encolhendo suas zonas de crescimento e conforto
